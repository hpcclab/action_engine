### Check the API name with invalid accessor is included  and remove all the invalid systax.
There are invalid syntax data and irregular accessor in the data
Ex. Altough the output of {'name': 'BookMovie', 'value': ['time', 'location']} is ust an string it has an accessor of .time and .location


```python
import ast
import re
from collections import defaultdict

def standardize_quotes(input_string):
    # This regex finds single quotes within arguments like 'Jenny' or 'Chill Vibes'
    # and replaces them with double quotes to avoid parsing errors
    return re.sub(r"(?<=\w)\'(?=\w)|(?<=\w)\'(?=\s)", '"', input_string)

# AST Node Visitor class to find function calls followed by attribute access
class DotAccessParser(ast.NodeVisitor):
    def __init__(self):
        # Dictionary to store API names with dot-accessed attributes
        self.api_with_dot_access = defaultdict(set)

    def visit_Attribute(self, node):
        # Check if the attribute is accessed on a function call
        if isinstance(node.value, ast.Call) and isinstance(node.value.func, ast.Name):
            api_name = node.value.func.id
            attribute_name = node.attr
            self.api_with_dot_access[api_name].add(attribute_name)  # Store the attribute
        self.generic_visit(node)


# Initialize parser and counters
parser = DotAccessParser()
failed_indices = []
delete_count = 0
total_items_checked = len(data)

# Process each item in data
for d in data[:]:  # Iterate over a copy of the list to modify it safely
    function_call = standardize_quotes(d["Label"])
    try:
        # Attempt to parse the function call
        tree = ast.parse(function_call)
        parser.visit(tree)
    except SyntaxError:
        # Record the index and remove the item if there is a SyntaxError
        failed_indices.append(d["Index"])
        data.remove(d)
        delete_count += 1

# Convert the result to the required format
apis_with_dot_access = [
    {"name": api, "value": list(attributes)}
    for api, attributes in parser.api_with_dot_access.items()
]

# Calculate remaining items
remaining_items = total_items_checked - delete_count

# Output results
print("APIs with dot-accessed outputs:", apis_with_dot_access)
print("Total number of items checked:", total_items_checked)
print("Number of items removed due to errors:", delete_count)
print("Indices of removed items:", failed_indices)
print("Number of items remaining:", remaining_items)
```

```python
import ast
import os
from ruamel.yaml import YAML
from collections import defaultdict

# Helper to parse dependencies and parameters
class DependencyParser(ast.NodeVisitor):
    def __init__(self):
        self.dependencies = defaultdict(set)  # Using set for unique dependencies
        self.parameters = defaultdict(dict)   # Using dict for unique parameters per name
        self.current_parent = None

    def visit_Call(self, node):
        func_name = node.func.id if isinstance(node.func, ast.Name) else None
        if func_name:
            if self.current_parent:
                # Record function dependency if it's not already added
                self.dependencies[self.current_parent].add(func_name)
            self.current_parent = func_name
            
            # Handle arguments and differentiate dependencies
            for kw in node.keywords:
                param_name = kw.arg
                param_value = kw.value

                # Check if the parameter is a nested function (indicating dependency)
                if isinstance(param_value, ast.Call):
                    nested_func_name = param_value.func.id if isinstance(param_value.func, ast.Name) else None
                    if nested_func_name:
                        # Add dependent parameter
                        self.parameters[func_name][param_name] = {
                            'name': param_name,
                            'value': f"{{{{ tasks.{nested_func_name.lower()}.result }}}}"
                        }
                        self.visit(param_value)  # Recursive call to check further nesting
                else:
                    # Add parameter as user input
                    self.parameters[func_name][param_name] = {
                        'name': param_name,
                        'value': f"{{{{ inputs.parameters.{param_name} }}}}"
                    }

# Initialize YAML handler
yaml = YAML()
yaml.default_flow_style = False

# Directory mapping for levels
level_directories = {
    "level1": level1,
    "level2": level2,
    "level3": level3
}

# Process each level
for level_name, level_data in level_directories.items():
    # Ensure the directory exists
    os.makedirs(level_name, exist_ok=True)
    
    # Process each item in the level
    for item in level_data:
        label = item["Label"]
        index = item["Index"]
        
        # Generate function call and parse
        function_call = label
        parser = DependencyParser()
        try:
            tree = ast.parse(function_call)
            parser.visit(tree)
            
            # Generate tasks based on dependencies and parameters
            tasks = []
            for node in reversed(parser.dependencies):  # Reverse the order of nodes
                dependencies = parser.dependencies[node]
                task = {
                    'name': node.lower(),
                    'template': node.lower(),
                }
                if dependencies:
                    task['dependencies'] = [dep.lower() for dep in dependencies]  # Convert each dependency to lowercase
                
                # Add parameters to each task
                task['arguments'] = {
                    'parameters': list(parser.parameters[node].values())  # Use unique parameters only
                }
                tasks.append(task)
            
            # Create the YAML structure
            argo_template = {
                'apiVersion': 'argoproj.io/v1alpha1',
                'kind': 'Workflow',
                'metadata': {
                    'generateName': f'dependency-workflow-{label}-'
                },
                'spec': {
                    'entrypoint': 'main',
                    'templates': [
                        {
                            'name': 'main',
                            'dag': {
                                'tasks': tasks
                            }
                        }
                    ]
                }
            }

            # Save the YAML file
            file_name = f"test_yaml_{label}.yaml"
            file_path = os.path.join(level_name, file_name)
            with open(file_path, 'w') as file:
                yaml.dump(argo_template, file)
            
            print(f"Generated YAML file for {label} at {file_path}")
        
        except SyntaxError:
            print(f"Syntax error in function call for item at {level_name}[{index}] - {label}")
```

```python
import json
# Example usage
file_path1 = "/home/UNT/ae0589/project/action_engine/eval/data/test_query/level1_queries.json"
file_path2 = "/home/UNT/ae0589/project/action_engine/eval/data/test_query/level2_queries.json"
file_path3 = "/home/UNT/ae0589/project/action_engine/eval/data/test_query/level3_queries.json"

# output_path = './eval/data/parsed_data.json'  # Path to save the parsed data

def load_dataset(file_path):
    """
    Load the dataset from a JSON file.
    """
    with open(file_path, 'r') as f:
        return json.load(f)

data1 = load_dataset(file_path1)
data2 = load_dataset(file_path2)
data3 = load_dataset(file_path3)
```

```python
# Define the function to calculate average, max, and min length and word count
def query_statistics(data):
    lengths = [len(entry["Query"]) for entry in data]
    word_counts = [len(entry["Query"].split()) for entry in data]
    
    average_length = sum(lengths) / len(data)
    average_word_count = sum(word_counts) / len(data)
    
    max_length = max(lengths)
    min_length = min(lengths)
    
    max_word_count = max(word_counts)
    min_word_count = min(word_counts)
    
    return average_length, average_word_count, max_length, min_length, max_word_count, min_word_count

# Calculate statistics for each dataset
stats_data1 = query_statistics(data1)
stats_data2 = query_statistics(data2)
stats_data3 = query_statistics(data3)

# Print the results
print(f"Data1 - Average length: {stats_data1[0]}, Average word count: {stats_data1[1]}, Max length: {stats_data1[2]}, Min length: {stats_data1[3]}, Max word count: {stats_data1[4]}, Min word count: {stats_data1[5]}")
print(f"Data2 - Average length: {stats_data2[0]}, Average word count: {stats_data2[1]}, Max length: {stats_data2[2]}, Min length: {stats_data2[3]}, Max word count: {stats_data2[4]}, Min word count: {stats_data2[5]}")
print(f"Data3 - Average length: {stats_data3[0]}, Average word count: {stats_data3[1]}, Max length: {stats_data3[2]}, Min length: {stats_data3[3]}, Max word count: {stats_data3[4]}, Min word count: {stats_data3[5]}")
```
